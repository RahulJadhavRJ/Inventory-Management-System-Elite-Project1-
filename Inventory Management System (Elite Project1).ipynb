```python
from tkinter import *
import tkinter

top = Tk()
top.geometry('200x200')
fd = open("store.txt","r")
txt = fd.read()
fd.close()
products = txt.split("\n")
Lb1 = Listbox(top)
for product in products:
    prod = product.split(",")
    Lb1.insert(1,prod)


Lb1.pack()
top.mainloop()


fd = open("store.txt","r")
txt = fd.read()
fd.close()

products = txt.split("\n")
prod_num=int(input("Enter how many products you want to buy :"))
ui_prod_list=[]
ui_quant_list=[]
total_amount=[]
    
for j in range(0,prod_num):
    ui_prod_list.append(input("Enter the product_Id: "))
    ui_quant_list.append(input("Enter the quantity: "))
    
   
for product in products:
    prod = product.split(",")
    
    for q in range(0,prod_num): 
       
        if(prod[0] == ui_prod_list[q]):
            print("***********************")
            print("Product ID: ", prod[0])
            print("Name: ", prod[1])
            print("Price: ",prod[2])
            print("Quant: ",ui_quant_list[0])
            print("------------------------")
            print("Billing Amout: ", int(ui_quant_list[q]) * int(prod[2]))
            total_amount.append(int(ui_quant_list[q]) * int(prod[2]))
            print("***********************")
            
print("list total :",total_amount)
print("***********************")
print("Total Amount to Pay :",sum(total_amount))   
print("***********************")

    
new_record = []

for product in products:
    prod = product.split(",")
    
    for s in range(0,prod_num): 
        if(ui_prod_list[s] == prod[0]):
            prod[3] = str(int(prod[3]) - int(ui_quant_list[s]))        
    new_record.append(prod[0] + "," + prod[1] + "," + prod[2] + "," + prod[3] + "\n")
    
new_record[-1] = new_record[-1][:-1]



fd = open("store.txt", 'w')

for i in new_record:
    fd.write(i)
    
fd.close()


new_record2 = []
fd = open("sell.txt","r")
txt = fd.read()
fd.close()

products2 = txt.split("\n")

for product2 in products2:
    prod = product2.split(",")  
    
    for r in range(0,prod_num): 
        if(ui_prod_list[r] == prod[0]):
            prod[3] = str(int(prod[3]) + int(ui_quant_list[r]))       
    new_record2.append(prod[0] + "," + prod[1] + "," + prod[2] + "," + prod[3] + "\n")
    
new_record2[-1] = new_record2[-1][:-1]



fd = open("sell.txt", 'w')

for i in new_record2:
    fd.write(i)
    
fd.close()
```
